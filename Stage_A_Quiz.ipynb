```python
df = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='latin-1')
df
```

```python
dfE = df.groupby('Element').sum()
dfE.sum()
```

```python
select2017Area = df[['Y2017', 'Area']]
grouped2017Area = select2017Area.groupby('Area').sum()
sorted_grouped2017Area = grouped2017Area.sort_values('Y2017')
seventh_lowest_area = sorted_grouped2017Area.index[6]
print("Area with the 7th lowest sum in 2017:", seventh_lowest_area)
```

```python
groupedE = df.groupby('Element').sum()
groupedE
```

```python
columns_2014 = [col for col in df.columns if col.endswith('2014')]
total_missing = df[columns_2014].isnull().sum().sum()
total_data_points = df[columns_2014].size
percentage_missing = (total_missing / total_data_points) * 100
print("Total number of missing data in 2014:", total_missing)
print("Percentage of missing data in 2014:", round(percentage_missing, 3))
```

```python
grouped = df.groupby('Area')['Y2017'].sum()
area_with_highest_sum = grouped.idxmax()
print("Area with the highest sum in 2017:", area_with_highest_sum)
```

```python

df['Y2017'] = df['Y2017'].astype(float)  
mean_2017 = df['Y2017'].mean()
std_2017 = df['Y2017'].std()

mean_2017_rounded = round(mean_2017, 2)
std_2017_rounded = round(std_2017, 2)

print("Mean for 2017:", mean_2017_rounded)
print("Standard Deviation for 2017:", std_2017_rounded)
```
